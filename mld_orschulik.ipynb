### Rearrange the dataframe
Here we create a function to rearrange our pandas dataframe to a dataframe with "segemted_file" as index.
Inside the function we dropped all colums, but the "segmented_file"-column and set it as index.
All the segmented files are inside the new folder "all_wavs"  that is set as a new audio directory.

```python
df_q = pd.read_pickle('./processed/Orschulik/Orschulik_mdf.pkl')
def re_arrange_df(df, audio_dir):
    df['segmented_file'] = df['segmented_file'].apply(lambda x: audio_dir+os.path.basename(x))
    df = df.set_index(df.segmented_file)
    df = df.drop(columns=['segmented_file', 'channel'])
    return df
df_q = re_arrange_df(df_q, 'all_wavs/')
```

### We set up a feature extractor for low level descriptors and functionals of a pre-defined feature set

```python
smile_lld = opensmile.Smile(
    feature_set = opensmile.FeatureSet.GeMAPSv01b,
    feature_level = opensmile.FeatureLevel.LowLevelDescriptors,
)
smile_functionals = opensmile.Smile(
    feature_set = opensmile.FeatureSet.GeMAPSv01b,
    feature_level = opensmile.FeatureLevel.Functionals,
)
```

```python
lld_df = smile_lld.process_files(df_q.index)
print(lld_df.shape)
```

```python
funct_df = smile_functionals.process_files(df_q.index)
print(funct_df.shape)
```

The low level descriptors describe each 10 ms of the original audio file. The low level descriptors dataframe has the dimensionality: 3526 rows and 18 columns.

```python
funct_df.head(3)
```

The functionals dataframe consists of 13 rows because of 13 segmented_files in which the original audio file was segmented. Each segment is described by a vector of fixed dimensionality after applying the following functionals:

* Extreme values and positions

* Means (arithmetic, quadratic, geometric)

* Moments (standard deviation, variance, kurtosis, skewness)

* Percentiles and percentile ranges

* Regression (linear and quadratic approximation, regression error)

* Centroid

* Peaks

* Segments

* Sample values

* Times/durations

* Onsets/Offsets

* Discrete Cosine Transformation (DCT)

* Zero Crossings

* Linear Predictive Coding (LPC) coefficients and gain


The functionals dataframe has the dimensionality: 13 rows and 62 columns.
